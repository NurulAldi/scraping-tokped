```python
import pandas as pd
from datetime import datetime
import re
```

```python
df = pd.read_json("../data/raw/hasil_tokopedia_scrap.json")
print("Jumlah data mentah: ", len(df))
df.head()
```

```python
df.info()
df.describe(include='all')
```

```python
df.columns = ['nama_produk', 'harga', 'rating', 'nama_toko', 'lokasi_toko', 'url']
```

```python
df.head(1)
```

```python
def clean_price(harga):
    if isinstance(harga, str):
        harga = re.sub(r'[^0-9]', '', harga)
        return int(harga) if harga else None
    return None
```

```python
df['harga'] = df['harga'].apply(clean_price)
```

```python
df.head()
```

```python
before = len(df)
df = df.drop_duplicates(subset=['nama_produk', 'nama_toko'])
after = len(df)
print(f"Duplikat dihapus: {before - after} baris")
```

```python
df['timestamp_cleaned'] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
```

```python
df = df.sort_values(by='harga', ascending=True).reset_index(drop=True)
```

```python
output_path = "../data/cleaned/tokopedia_headphone_cleaned.csv"
df.to_csv(output_path, index=False)
print(f"Data bersih disimpan ke {output_path}")
```
